```python
splitted = tmp.split('/')
print('oid: ', splitted[-2])
print('aid: ', splitted[-1])
```

```python
oid = splitted[-2]
aid = splitted[-1]
```

```python
import requests
from requests import H

# res = requests.get(f'https://m.entertain.naver.com/api/article/getArticle?oid={oid}&aid={aid}')
res = requests.get(tmp)
raw = res.content
html = raw.decode('utf-8')
html
```

```python
from bs4 import BeautifulSoup as bs

soup = bs(html, 'html.parser')
soup
```

```python
title_tag = soup.find("div", class_="ArticleHead_article_head_title__YUNFf")
content_tag = soup.find("div", class_="_article_content")

print('title: ', title_tag)
print('content: ', content_tag)
```
